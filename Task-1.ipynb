```python
class Student:
    def __init__(self, name, student_id, major):
        """
        Initializes a Student object with name, student ID, and major.
        """
        self.name = name
        self.student_id = student_id
        self.major = major

    def get_details(self):
        """
        Returns a string with the student's details.
        """
        return f"Name: {self.name}, Student ID: {self.student_id}, Major: {self.major}"

# Create an instance of the Student class
student1 = Student("Alice Smith", "S12345", "Computer Science")

# Display the student's details
print(student1.get_details())
```
